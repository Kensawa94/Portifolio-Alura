```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker     
```

```
data1 = pd.read_csv('A151346189_28_143_208.csv', decimal = ',' ,skiprows=3, skipfooter = 12 ,delimiter = ';', encoding = 'iso-8859-1',engine='python')
data1 = data1.replace('-', 0.0)
data1['2020/Ago'] = data1['2020/Ago'].astype('float64')
pd.options.display.float_format = '{:.2f}'.format
```

```
data1.dtypes #Validação de todas as colunas já colocadas como float64
```

```
data1.plot(x='Unidade da Federação', y = '2008/Ago', kind ='bar', figsize=(9, 7))
```

```
ax = data1.plot(x='Unidade da Federação', y = '2008/Ago', kind ='bar', figsize=(9, 7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R${x:,.2f}'))
plt.title("Custo de Procedimentos Hospitalares por Unidade de Federação (Estado)")
plt.show()
```

```
#gráfico CRU
ax = data1.plot(x='Unidade da Federação', y = ('2020/Ago'), kind ='bar', figsize=(9, 7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R${x:,.2f}'))
plt.title("Custo de Procedimentos Hospitalares por Unidade de Federação (Estado) - ")
plt.grid()
plt.show()
```

```
#gráfico divisão do valor lateral
plt.bar(data1['Unidade da Federação'], data1['2020/Ago'], width=0.8, bottom=None, align='center')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R${x:,.2f}'))
plt.title("Custo de Procedimentos Hospitalares por Unidade de Federação (Estado) - ")
plt.grid()
plt.show()
```
